```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('new_retail_data.csv')

cols = ["Transaction_ID", "Customer_ID", "Zipcode", "Ratings", "Total_Purchases", 'Age']
df[cols] = df[cols].astype("Int64")

df.head()
```

```python
#Remove customers that have different emails for the same customer id
valid_ids = (
    df.groupby("Customer_ID")["Email"]
    .nunique()
    .eq(1)
)

df_clean = df[df["Customer_ID"].isin(valid_ids[valid_ids].index)]

df = df_clean.copy()

df_clean.head()

print(len(df))
print(len(df_clean))
```

```python
customer_info_cols = ['Customer_ID', 'Name', 'Email', 'Phone', 'Address', 'City', 'State', 'Zipcode', 'Country', 'Age', 'Gender', 'Income', 'Customer_Segment']
```

```python
#Save non-customer columns as transaction data columns
df_transactions = df.drop(columns=customer_info_cols[1:])

# Make transaction date and time into one timestamp column
df_transactions['Transaction_Date'] = pd.to_datetime(df_transactions['Date'] + ' ' + df_transactions['Time'])

#Also drop out Year, Month, Date columns
df_transactions = df_transactions.drop(columns=['Date', 'Time', 'Year', 'Month'])

#Show final transactions dataframe
df_transactions.head()
```

```python
# Rearrange column order

# Specify the column to move and the desired new position (index)
column_to_move = 'Transaction_Date'
new_position = 2 # To move to the first position

# 1. Use pop() to remove the column and store its data
col_data = df_transactions.pop(column_to_move)

# 2. Use insert() to place the column at the desired position
df_transactions.insert(new_position, column_to_move, col_data)

print("New DataFrame columns:", df_transactions.columns)

df_transactions.head()
```

```python
df_customer_info = df[customer_info_cols]

# Convert phone number from float to string
df_customer_info['Phone'] = df_customer_info['Phone'].fillna(0).astype('int64').astype('string')

df_customer_info.head()
```

```python
# Write separated dataframes to CSV
df_transactions.to_csv('retail_transactions.csv', index=False)
df_customer_info.to_csv('retail_customers.csv', index=False)
```
